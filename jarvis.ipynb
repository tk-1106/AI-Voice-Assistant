```python
import pyaudio
import pyttsx3
import datetime
import wikipedia
import smtplib # inbuild library for email
import webbrowser as wb
import speech_recognition as sr
import os
import pyautogui #!pip install pyautogui  for screen shot, mouse control,key board control,cross platform like mac,os,linux
import psutil #!pip install psutil --CPU, memory, disks, network, sensors)
import pyjokes #!pip install pyjokes

engine=pyttsx3.init()

def speak(audio,rate=150):
    
    engine.say(audio)
    engine.setProperty("rate",rate)
    engine.runAndWait()
    
def hi():
    welcome=("hi and hello karthick")
    speak(welcome)
#hi()
    
def time():
    Time = datetime.datetime.now().strftime("%I:%M:%S")
    speak("the current time is")
    speak(Time)
    
#time()

def date():
    
    Year=datetime.datetime.now().year
    Month=datetime.datetime.now().month
    Date=datetime.datetime.now().day
    speak("the current date is!")
    speak(Date)
    speak(Month)
    speak(Year)
    
#date()


def greet():

    hi()
    
    speak("welcome back karthick!")
    

    time()
    
    date()
    
    hour=datetime.datetime.now().hour
    if hour >=6 and hour < 12:
        speak("Good morning sir!")
    elif hour>=12 and hour <18:
        speak("Good afternoon sir!")
    elif hour >=18 and hour < 22:
        speak("Good evening sir")
    else:
        speak("Good night sir!")
    
    speak("hi i am karthick how can i help you ?")
    
#greet()

def quit():
    print("Exiting the program...")
    # Additional cleanup or finalization steps can be added here
# quit()

# so now install speech recognition
#!pip install SpeechRecognition
def takecommand():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        print("listening...")
        r.pause_threshold = 1
        audio = r.listen(source)
        
        
        try:
            
            print("recognizing..")
            query=r.recognize_google(audio , language="en-in")
            print(query)

            
        except Exception as e:
            print(e)
            speak("say that again please...!")
            return "None"
        return query

def sendEmail(to,content):
    server=smtplib.SMTP("smtp.gmail.com",587)
    server.ehlo()
    server.starttls()
    server.login("mrkarthick104@gmail.com","dolcymkwerxvqkgs")
    server.sendmail("karthekash@gmail.com.com",to,content)
    server.close()
#sendEmail(to,content)
#takecommand()

def screenshot():
    image= pyautogui.screenshot()
    image.save(r"C:\Users\Karthick\Documents\Data science\ai project\karthick programs\ss1.png")
#screenshot()
def cpu():
    usage= str(psutil.cpu_percent())
    speak("cpu rate is "+usage)
    battery=str(psutil.sensors_battery())
    speak("battery rate is"+battery)
    print(battery)
#cpu()
def jokes():
    joke=pyjokes.get_joke()
    print(joke)
    speak(joke)
#jokes()

if __name__ == "__main__":
    #greet()
    while True:
        query = takecommand().lower()
        
        if "time" in query:
            time()
        elif "date" in query:
            date()
        elif "wikipedia" in query:
            speak("Searching...")
            query = query.replace("wikipedia", "").strip()
            try:
                result = wikipedia.summary(query, sentences=3)
                print(result)
                speak(result)
            except wikipedia.exceptions.PageError:
                speak("Sorry, I couldn't find any information on that topic.")
        elif "send email" in query:
            try:
                speak("what i want to say..?")
                content=takecommand()
                to="karthekash@gmail.com" # use any gmail account
                sendEmail(to,content)
                speak(content)
                speak("mail is sent")
            except Exception as e:
                print(e)
                speak("unable to sent email")
        elif "search in chrome" in query:
            speak("What should I search?")
            search = takecommand().lower()
            chromepath = 'C:/Program Files/Google/Chrome/Application/chrome.exe %s'  # Use forward slashes for file paths
            wb.get(chromepath).open_new_tab('https://www.google.com/search?q=' + search)
        elif "logout" in query:
            os.system("shutdown -l")
        elif "shutdown" in query:
            os.system("shutdown /s /t 1")
        elif "restart" in query:
            os.system("shutdown /r /t 1")
        elif "play songs" in query:
            song_path="E:\\song"
            songs=os.listdir(song_path)
            if songs:
                #random_song = random.choice(songs)
                os.startfile(os.path.join(song_path,songs[0])) # or(   os.startfile(os.path.join(song_path,random_songs) 
            else:
                speak("there is no songs in directory")
                print("there is no songs in directory")
        elif "remember that" in query:
            speak("what should i want to remember")
            data = takecommand()
            speak("you said me what me want to remember"+data)
            remember = open("remi.txt","w")
            remember.write(data)
            remember.close()
        elif "what i previously said" in query:
            remember=open("remi.txt","r")
            speak("u previously  said that"+remember.read())
            remember.close()            
        elif "take screenshot" in query:
            screenshot()
            speak("Done")
        elif "cpu" in query:
            speak("your cpu and battery rate is")
            cpu()
        elif "joke" in query:
            jokes()
        elif "offline" in query:
            speak("yes")
            speak("thank you for your interaction")
            quit()
            break
#pip install wikipedia
    


        
        
    
```
